```python
import random
def calcular(retweets):
    ganancias = retweets * random.randint(3, 5)
    return ganancias

df["ganancias"] = df["retweets"].apply(calcular)
df.head(10)


def popularida(fila):
    resultado = fila["followers"]/fila["followers"]
    return resultado

df["popularidad"] = df.apply(popularida, axis=1)
df.head()
```

```python
grupo = df.groupby("country").agg({
    "followers":'sum',
    "mentions":'mean',
    "retweets":'max'
})

grupo[grupo["followers"]>5000]
```

```python
import matplotlib.pyplot as grafi


grupo["followers"].plot()
grafi.show()
```
